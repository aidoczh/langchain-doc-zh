```markdown
# 分析单个长文档

AnalyzeDocumentChain接收单个文档，将其拆分，并通过CombineDocumentsChain进行处理。
```

````python
with open("../docs/docs/modules/state_of_the_union.txt") as f:
    state_of_the_union = f.read()
````

````python
# 导入AnalyzeDocumentChain类
from langchain.chains import AnalyzeDocumentChain
# 导入ChatOpenAI类
from langchain_openai import ChatOpenAI

# 创建ChatOpenAI实例，并指定模型为"gpt-3.5-turbo"，温度为0
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
````

````python
从langchain.chains.question_answering导入load_qa_chain函数

# 载入问答链
qa_chain = load_qa_chain(llm, chain_type="map_reduce")
````

````python
qa_document_chain = AnalyzeDocumentChain(combine_docs_chain=qa_chain)
# 创建一个名为qa_document_chain的变量，用于存储通过AnalyzeDocumentChain类创建的对象
# 该对象的combine_docs_chain参数设置为qa_chain变量的值
````

````python
# 导入qa_document_chain模块
import qa_document_chain

# 运行qa_document_chain的run函数
qa_document_chain.run(
    input_document=state_of_the_union,  # 输入参数：state_of_the_union，表示输入的文档
    question="what did the president say about justice breyer?",  # 输入参数：question，表示问题的内容
)
````
