# 伪造的LLM
LangChain提供了一个伪造的LLM类，用于测试。这样可以模拟对LLM的调用，并模拟LLM以特定方式响应的情况。

在这个笔记本中，我们将介绍如何使用它。

我们首先使用FakeLLM在一个代理中开始。

```python

# 导入FakeListLLM模块
from langchain_community.llms.fake import FakeListLLM
```

```python
# 导入所需模块
from langchain.agents import AgentType, initialize_agent, load_tools
```

```python
# 调用load_tools函数，传入参数["python_repl"]，返回结果赋值给变量tools
tools = load_tools(["python_repl"])
```

```python
# 定义一个包含两个元素的列表responses
responses = ["Action: Python REPL\nAction Input: print(2 + 2)", "Final Answer: 4"]

# 创建一个FakeListLLM对象，并将responses作为参数传递给构造函数
llm = FakeListLLM(responses=responses)
```

```python
# 初始化代理
agent = initialize_agent(
    tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True
)
```

```python
# 使用agent对象调用invoke方法，传入参数"whats 2 + 2"
agent.invoke("whats 2 + 2")
```
