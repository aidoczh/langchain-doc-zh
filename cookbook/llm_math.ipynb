# 数学链这个笔记本展示了如何使用LLMs和Python REPL来解决复杂的单词数学问题。

```python
# 导入所需的库from langchain.chains import LLMMathChainfrom langchain_openai import OpenAI# 创建一个OpenAI实例，设置温度为0llm = OpenAI(temperature=0)# 从OpenAI实例创建一个LLMMathChain实例，并设置为详细模式llm_math = LLMMathChain.from_llm(llm, verbose=True)# 调用LLMMathChain实例的invoke方法，询问“13的.3432次方是多少？”llm_math.invoke("What is 13 raised to the .3432 power?")
```
