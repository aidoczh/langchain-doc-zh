# 计划和执行

计划和执行代理通过首先规划要做什么，然后执行子任务来实现目标。这个想法在很大程度上受到[BabyAGI](https://github.com/yoheinakajima/babyagi)的启发，然后是["Plan-and-Solve" paper](https://arxiv.org/abs/2305.04091)。

规划几乎总是由LLM完成。

执行通常由一个单独的代理完成（配备工具）。

## 导入模块

```python
# 导入所需的模块和类
from langchain.chains import LLMMathChain
from langchain_community.utilities import DuckDuckGoSearchAPIWrapper
from langchain_core.tools import Tool
from langchain_experimental.plan_and_execute import (
    PlanAndExecute,
    load_agent_executor,
    load_chat_planner,
)
from langchain_openai import ChatOpenAI, OpenAI
```

## 工具

```python
# 导入所需的库
from DuckDuckGoSearchAPIWrapper import DuckDuckGoSearchAPIWrapper
from OpenAI import OpenAI
from LLMMathChain import LLMMathChain

# 创建DuckDuckGoSearchAPIWrapper对象
search = DuckDuckGoSearchAPIWrapper()

# 创建OpenAI对象，并设置temperature为0
llm = OpenAI(temperature=0)

# 使用LLM创建LLMMathChain对象，并设置verbose为True
llm_math_chain = LLMMathChain.from_llm(llm=llm, verbose=True)

# 创建工具列表
tools = [
    Tool(
        name="Search",  # 工具名称为Search
        func=search.run,  # 工具功能为运行search的run方法
        description="useful for when you need to answer questions about current events",  # 工具描述为回答有关当前事件的问题时很有用
    ),
    Tool(
        name="Calculator",  # 工具名称为Calculator
        func=llm_math_chain.run,  # 工具功能为运行llm_math_chain的run方法
        description="useful for when you need to answer questions about math",  # 工具描述为回答有关数学问题时很有用
    ),
]
```
