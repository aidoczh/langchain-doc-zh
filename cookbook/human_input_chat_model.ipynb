# 人类输入聊天模型

除了HumanInputLLM之外，LangChain还提供了一个伪聊天模型类，可用于测试、调试或教育目的。这使您可以模拟对聊天模型的调用，并模拟人类在接收到消息后如何回复。

在这个笔记本中，我们将介绍如何使用它。

我们首先使用HumanInputChatModel在一个代理中开始。

```python
# 导入HumanInputChatModel类
from langchain_community.chat_models.human import HumanInputChatModel
```

由于我们将在这个笔记本中使用`WikipediaQueryRun`工具，如果您还没有安装`wikipedia`包，您可能需要安装它。

```python
# 安装wikipedia库
%pip install wikipedia
```

```python
# 导入所需模块
from langchain.agents import AgentType, initialize_agent, load_tools
```

```python

# 加载所需的工具
tools = load_tools(["wikipedia"])

# 创建一个人工输入聊天模型
llm = HumanInputChatModel()
```

```python
# 初始化一个代理
agent = initialize_agent(
    tools, llm, agent=AgentType.CHAT_ZERO_SHOT_REACT_DESCRIPTION, verbose=True
)
```

```python
# 调用agent函数，并传入字符串参数"What is Bocchi the Rock?"
agent("Bocchi the Rock是什么？")
```
