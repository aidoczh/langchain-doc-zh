# 人类输入LLM

与虚假LLM类似，LangChain提供了一个伪LLM类，可用于测试、调试或教育目的。这使您可以模拟对LLM的调用，并模拟人类在收到提示时如何响应。

在这个笔记本中，我们将介绍如何使用它。

我们首先使用HumanInputLLM在一个代理中。

```python
from langchain_community.llms.human import HumanInputLLM
```

```python
# 导入 langchain.agents 模块中的 AgentType 类、initialize_agent 函数和 load_tools 函数
from langchain.agents import AgentType, initialize_agent, load_tools
```

由于我们将在这个笔记本中使用`WikipediaQueryRun`工具，如果您还没有安装`wikipedia`包，您可能需要安装它。

```python
# 导入wikipedia库
%pip install wikipedia
```

```python
# 加载所需的工具
tools = load_tools(["wikipedia"])

# 创建一个人工输入的语言模型（LLM）
llm = HumanInputLLM(
    prompt_func=lambda prompt: print(
        f"\n===PROMPT====\n{prompt}\n=====END OF PROMPT======"
    )
)
```

```python
# 初始化代理
agent = initialize_agent(
    tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True
)
```

```python
# 运行代理程序，并传入参数"What is 'Bocchi the Rock!'?"
agent.run("What is 'Bocchi the Rock!'?")
```
